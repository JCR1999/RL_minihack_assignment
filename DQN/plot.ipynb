```python
#%%
# importing modules
import numpy as np
import matplotlib.pyplot as plt
  
# import sys
# sys.stdin = open('input.txt', 'r')
with open('text.txt', 'r') as file:
     input_lines = [line.strip() for line in file]
        
data = input_lines[1].split(',')
# print(data)
dqn_average_rewards_per_eps = []
for i in data:
    dqn_average_rewards_per_eps.append(float(i))

plt.plot(dqn_average_rewards_per_eps)

plt.xlabel("Episode")
plt.ylabel("Average reward")
plt.title("DQN reward per episode (averaged over 20 runs)")
  
# Function add a legend  
plt.legend(["DQN"], loc ="lower right")
  
plt.savefig('dqnReward.png', bbox_inches='tight')
# function to show the plot
plt.show()
# %%
```
